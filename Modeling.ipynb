## Transform Datasets

```python
#Features-Target Split
X_raw_full_df = raw_full_df.iloc[:,1:].copy()
Y_full_df = raw_full_df.iloc[:,0].copy()

X_raw_train_df = raw_train_df.iloc[:,1:].copy()
Y_train_df = raw_train_df.iloc[:,0].copy()

X_raw_test_df = raw_test_df.iloc[:,1:].copy()
Y_test_df = raw_test_df.iloc[:,0].copy()

#Feature Preprocessing: Normalizing and Centering Coordinates around the (X0, Y0) landmark
X_preprocessesd_full_df = transform_fn(X_raw_full_df)
X_preprocessesd_train_df = transform_fn(X_raw_train_df)
X_preprocessesd_test_df = transform_fn(X_raw_test_df)
```

```python
X_preprocessesd_full_df.head()
```

## Train Model

```python
from sklearn.ensemble import RandomForestClassifier
```

### Full Dataset, Random Split:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_preprocessesd_full_df.copy(), Y_full_df.copy(), 
                                                    test_size=0.2, shuffle=True, stratify = Y_full_df.copy())
model0 = RandomForestClassifier()

model0.fit(X_train, y_train)
y_pred = model0.predict(X_test)
score = accuracy_score(y_test, y_pred)*100
print(score)
```

### Train and Test obtained individually, not splitted:

```python
model1 = RandomForestClassifier()

model1.fit(X_preprocessesd_train_df, Y_train_df)
y_pred1 = model1.predict(X_preprocessesd_test_df)
score = accuracy_score(Y_test_df, y_pred1)*100
print(score)
```
